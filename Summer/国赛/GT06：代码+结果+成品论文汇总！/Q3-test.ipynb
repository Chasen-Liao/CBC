```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
plt.rcParams['font.sans-serif'] = ['KaiTi']

def analyze_reflectance_data(file_path):
    # 读取数据
    try:
        df = pd.read_excel(file_path)
    except FileNotFoundError:
        print(f"文件 '{file_path}' 未找到，请检查文件路径。")
        return
    except Exception as e:
        print(f"读取文件时发生错误: {e}")
        return

    # 确保数据中包含波数和反射率列
    if '波数 (cm-1)' not in df.columns or '反射率 (%)' not in df.columns:
        print("数据文件中缺少必要的列（波数 (cm-1) 或 反射率 (%)）")
        return

    # 提取波数和反射率数据
    wavenumbers = df['波数 (cm-1)'].values
    reflectances = df['反射率 (%)'].values

    # 查找反射率数据中的峰值
    peaks, _ = find_peaks(reflectances, height=0)

    # 绘制反射率随波数的变化图
    plt.figure(figsize=(12, 6))
    plt.plot(wavenumbers, reflectances, label='反射率')
    plt.scatter(wavenumbers[peaks], reflectances[peaks], color='red', label='反射率峰值', marker='o')
    plt.xlabel('波数 (cm-1)')
    plt.ylabel('反射率 (%)')
    plt.title('附件3反射率随波数的变化')
    plt.legend()
    plt.grid(True)
    plt.savefig('附件3反射率随波数的变化.png', dpi=300, bbox_inches='tight')
    plt.show()

    # 简单判断是否可能存在多光束干涉
    if len(peaks) > 0:
        print("反射率数据中存在峰值，可能存在多光束干涉现象，需要进一步结合实际情况判断。")
        # 计算相邻峰值之间的波数差
        peak_wavenumber_diffs = np.diff(wavenumbers[peaks])
        print("相邻反射率峰值之间的波数差:", peak_wavenumber_diffs)
    else:
        print("反射率数据中未检测到明显峰值，多光束干涉现象不明显或不存在。")


# 分析附件3数据
analyze_reflectance_data('附件3.xlsx')
```
