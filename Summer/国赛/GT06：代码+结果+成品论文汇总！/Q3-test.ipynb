```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
plt.rcParams['font.sans-serif'] = ['KaiTi']

def analyze_reflectance_data(file_path):
    # 
    try:
        df = pd.read_excel(file_path)
    except FileNotFoundError:
        print(f" '{file_path}' ")
        return
    except Exception as e:
        print(f": {e}")
        return

    # 
    if ' (cm-1)' not in df.columns or ' (%)' not in df.columns:
        print(" (cm-1)   (%)")
        return

    # 
    wavenumbers = df[' (cm-1)'].values
    reflectances = df[' (%)'].values

    # 
    peaks, _ = find_peaks(reflectances, height=0)

    # 
    plt.figure(figsize=(12, 6))
    plt.plot(wavenumbers, reflectances, label='')
    plt.scatter(wavenumbers[peaks], reflectances[peaks], color='red', label='', marker='o')
    plt.xlabel(' (cm-1)')
    plt.ylabel(' (%)')
    plt.title('3')
    plt.legend()
    plt.grid(True)
    plt.savefig('3.png', dpi=300, bbox_inches='tight')
    plt.show()

    # 
    if len(peaks) > 0:
        print("")
        # 
        peak_wavenumber_diffs = np.diff(wavenumbers[peaks])
        print(":", peak_wavenumber_diffs)
    else:
        print("")


# 3
analyze_reflectance_data('3.xlsx')
```
