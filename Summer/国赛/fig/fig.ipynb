```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['font.sans-serif'] = ['WenQuanYi Zen Hei']

df = pd.read_excel("附件1.xlsx")
df
```

```python
# 设置图片清晰度
plt.rcParams['figure.dpi'] = 300

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 中文支持
plt.rcParams['axes.unicode_minus'] = False  # 修复负号显示问题

# 绘制折线图
plt.figure(figsize=(10, 6))
plt.plot(df['波数 (cm-1)'], df['反射率 (%)'], label='反射率 (%)', color='blue')

# 添加图标题和轴标签
plt.title('波数与反射率的关系')
plt.xlabel('波数 (cm-1)')
plt.xticks(rotation=45)
plt.ylabel('反射率 (%)')
plt.gca().invert_xaxis() # x轴反过来，因为红外光谱通常是波数从大到小
# 添加图例
plt.legend()

# 显示网格
plt.grid(True)

# 显示图形
plt.show()
```

平滑算法 + 标记峰值

```python
df = pd.read_excel("附件1.xlsx")
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.signal import savgol_filter, find_peaks
plt.rcParams['figure.dpi'] = 300

# 中文支持
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False

x = df['波数 (cm-1)']
y = df['反射率 (%)']

#平滑处理，减少波动带来的假峰值
y_smooth = savgol_filter(y, window_length=15, polyorder=3)
peaks, _ = find_peaks(y_smooth, prominence=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='原始数据', color='blue', alpha=0.5)
plt.plot(x, y_smooth, label='平滑数据', color='red')
plt.scatter(x.iloc[peaks], y_smooth[peaks], color='pink', marker='^', s=60, label='峰值')

# 添加图标题和轴标签
plt.title('波数与反射率的关系')
plt.xlabel('波数 (cm-1)')
plt.xticks(rotation=45)
plt.ylabel('反射率 (%)')

# x轴反转（红外光谱常用）看很多论文都是反过来的，因为要从波长小的开始
plt.gca().invert_xaxis()

# 添加图例
plt.legend()

# 显示网格
plt.grid(True)

# 显示图形
plt.show()
```
